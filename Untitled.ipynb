Accesing empty columns using coalesce

```python
%%sql
SELECT product_category, product_desc_length
FROM products
WHERE COALESCE(product_category, '') != ''
ORDER BY product_desc_length DESC
LIMIT 10;
```

Using LIKE (case sensitive) and ILIKE (case insensitive)

```python
%%sql
SELECT DISTINCT(product_category)
FROM products
WHERE product_category LIKE 'fash%';
```

```python
%%sql
SELECT DISTINCT(product_category)
FROM products
WHERE product_category LIKE '%sh%';
```
