INNER JOIN

```python
%%sql
SELECT orders.customer_id, orders.order_status AS status,
       customers.customer_zipcode AS zipcode,
       customers.customer_city AS city,
       customers.customer_state as state
FROM orders JOIN customers
ON orders.customer_id  = customers.customer_id
LIMIT 10;
```

JOINING multiple tables

```python
%%sql
SELECT od.customer_id, od.order_status AS status,
       cm.customer_zipcode AS zipcode,
       cm.customer_city AS city,
       cm.customer_state as state,
       op.payment_type ,
       op.payment_value AS value
FROM orders AS od JOIN customers AS cm
     ON od.customer_id = cm.customer_id
JOIN order_payments AS op 
     ON od.order_id = op.order_id
WHERE cm.customer_state = 'SP'
LIMIT 10;
```

```python
%%sql
SELECT cm.customer_state as "state", 
       CAST(SUM(op.payment_value) AS numeric(20,2)) AS "value"
FROM orders AS od JOIN customers AS cm
     ON od.customer_id = cm.customer_id
JOIN order_payments AS op 
     ON od.order_id = op.order_id
GROUP BY cm.customer_state
ORDER by value DESC;
```

Advance Queries

Subquery using WHERE clause

```python
%%sql
SELECT order_item_id as id,
       shipping_limit_date,
       price,
       freight_value 
FROM order_items
WHERE freight_value > (SELECT percentile_cont(.9) WITHIN GROUP (ORDER BY freight_value)
                    FROM order_items)
LIMIT 10
```

Creating derived tables from subquery

```python
%%sql
SELECT category,
       ROUND(AVG(weight),2) AS avg_weight,
       ROUND(AVG(length),2) AS avg_length,
       ROUND(AVG(height),2) AS avg_height,
       ROUND(AVG(width),2) AS avg_width
FROM(SELECT product_category AS category,
       product_weight_grams AS weight,
       product_length_cm AS length,
       product_height_cm AS height,
       product_width_cm AS width
       FROM products) AS dimensions
GROUP BY category
```

Joining derived tables 

```python
%%sql
SELECT total_sales.seller_id,
       total_sales.total_sale,
       total_sales.total_units,
       (total_sales.total_sale/total_sales.total_units) AS avg_sales,
       sellers.city,
       sellers.state
FROM(SELECT seller_id , 
  SUM(order_item_id) AS total_units,
  SUM(price) AS total_sale
  FROM order_items
  GROUP BY seller_id) AS total_sales
JOIN( SELECT seller_id,
             seller_city AS city,
             seller_state AS state
      FROM sellers) AS sellers
ON total_sales.seller_id = sellers.seller_id
WHERE total_sales.total_units BETWEEN 1000 AND 2000
ORDER BY total_sales.total_sale DESC
LIMIT 10
```

Genrating column from queries

```python
%%sql 
SELECT payment_type,
       payment_installments,
       payment_value,
       payment_value - (SELECT percentile_cont(.5) WITHIN GROUP(ORDER BY payment_value)
        FROM order_payments) AS payment_deviation
FROM order_payments
LIMIT 10
```

Common Table Expressions

```python
%%sql
WITH 
    dimensions(category, weight, length, height, width) AS
    (SELECT product_category,
            product_weight_grams,
            product_length_cm,
            product_height_cm,
            product_width_cm
    FROM products)
SELECT category,
       ROUND(AVG(weight),2) AS avg_weight,
       ROUND(AVG(length),2) AS avg_length,
       ROUND(AVG(height),2) AS avg_height,
       ROUND(AVG(width),2) AS avg_width
FROM dimensions
GROUP BY category
LIMIT 10
```

```python
%%sql
WITH
    total_sales(seller_id,total_units,total_sale) AS
    (SELECT seller_id,
            SUM(order_item_id),
            SUM(price)
     FROM order_items
     GROUP BY seller_id),
    
    sellers(seller_id,city,state) AS
    (SELECT seller_id,
            seller_city,
            seller_state
    FROM sellers)

SELECT total_sales.seller_id,
       total_sales.total_sale,
       total_sales.total_units,
       (total_sales.total_sale/total_sales.total_units) AS avg_sales

FROM total_sales JOIN sellers
ON total_sales.seller_id = sellers.seller_id
WHERE total_sales.total_units BETWEEN 1000 AND 2000
ORDER BY total_sales.total_sale DESC
LIMIT 10
```

Reclassifying Values using CASE

```python
%%sql
SELECT order_id,
       price,
       freight_value,
       CASE WHEN freight_value = 0.0 THEN 'FREE'
            WHEN freight_value BETWEEN 0.1 AND 100.0 THEN 'LOW'
            WHEN freight_value BETWEEN 100.1 AND 200.0 THEN 'MEDIUM'
            ELSE 'HIGH'
       END AS freight_cost
FROM order_items
LIMIT 10
```
